```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

%matplotlib inline
```

```python
Ts = np.array([2, 6, 8, 9, 10, 11, 12, 15, 20, 30, 40])/10 #x axis
Cs = []
ss = []
rr = []
nn = []
tss = []
for N in [0,1,2,3,4]:
    for i, r in enumerate([4,5]):
        for size in [20, 40, 60, 80, 100, 120]:
            
            for T in [2, 6, 8, 9, 10, 11, 12, 15, 20, 30, 40]:
                temp_result = np.loadtxt('NewResults/r{}n{}T{}N{}E.txt'.format(r,N,T,size))[10000:] # tail -90000
                E = np.sum(temp_result)
                E2 = np.sum(temp_result * temp_result)
                n = temp_result.shape[0]
                C = (E2/n - E*E/n/n)/T/T*100
                Cs.append(C)
                ss.append(size)
                rr.append(r)
                nn.append(N)
                tss.append(T)
```

```python
Cs = pd.Series(Cs)
ss = pd.Series(ss)
rr = pd.Series(rr)
nn = pd.Series(nn)
tss = pd.Series(tss)
```

```python
df = pd.DataFrame({})
df['C'] = Cs
df['size'] = ss
df['r'] = rr
df['n'] = N
df['Temp'] = tss
```

```python
df['size'] = df['size'].apply(lambda x: 'N={}'.format(x))
```

```python
fig, ax = plt.subplots(1,1,figsize=(20,10))
sub_df = df[df['r'] == 4]
sns.lineplot(x='Temp',y='C',hue='size',data=sub_df, ax=ax)
plt.title("Specific Heat Smallworld, r=4, w0.4, N=0.3")
```

```python
fig, ax = plt.subplots(1,1,figsize=(20,10))
sub_df = df[df['r'] == 5]
sns.lineplot(x='Temp',y='C',hue='size',data=sub_df, ax=ax)
plt.title("Specific Heat Smallworld, r=5, w0.4, N=0.3")
```
