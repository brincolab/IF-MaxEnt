```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

%matplotlib inline
```

```python
fig, ax = plt.subplots(2,1,figsize=(10,10))

Ts = np.array([2, 6, 8, 9, 10, 11, 12, 15, 20, 30, 40])/10 #x axis

for i, r in enumerate([4,5]):
    for size in [20, 40, 60, 80, 100, 120]:
        Cs = []
        for T in [2, 6, 8, 9, 10, 11, 12, 15, 20, 30, 40]:
            temp_result = np.loadtxt('Results/r{}T{}N{}E.txt'.format(r,T,size))[10000:] # tail -90000
            E = np.sum(temp_result)
            E2 = np.sum(temp_result * temp_result)
            n = temp_result.shape[0]
            C = (E2/n - E*E/n/n)/T/T*100
            Cs.append(C)
        sns.lineplot(Ts,Cs,ax=ax[i])
        sns.scatterplot(Ts,Cs,ax=ax[i])



    ax[i].set_xlabel("$T$")
    ax[i].set_ylabel("$C(T)$")
    ax[i].set_title("Parametro r=0.{}".format(r))
plt.legend(['N=20','N=40','N=60','N=80','N=100','N=120'])
plt.show()
```
